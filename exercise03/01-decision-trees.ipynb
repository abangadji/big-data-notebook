## a)

Generate a decision tree to predict the tennis decision. Use the Gini-index to find the best splits.

$$
gini(T) = 1 - \sum_{j=1}^{k} p_j^2 \\
gini_A(T) = \sum_{i=1}^{m} \frac{\mid T_i \mid}{\mid T \mid} * gini(T_i)
$$

### I.

```
    (---outlook---)
  s/      o|      \r
[2+,3-] [4+,0-] [3+,2-]
 0.48     0       0.48
```

$ \Rightarrow gini(T,outlook) = \frac{5}{14}*0.48 + \frac{4}{14}*0 + \frac{5}{14} * 0.48 = 0.343 $

**Min:** outlook

---

```
    (--- temp ---)
  c/      m|      \h
[3,1]    [4,2]   [2,2]
0.375    0.444    0.5
```

$ \Rightarrow gini(T,temp) = \frac{4}{14}*0.375 + \frac{6}{14}*0.444 + \frac{4}{14} * 0.5 = 0.44 $

---

```
    (--- hum ---)
     h/       \n
   [3,4]     [6,1]
    0.5       0.245
```

$ \Rightarrow gini(T,hum) = 0.3725 $

---

```
    (--- windy ---)
     t/       \f
   [6,2]     [3,3]
   0.375      0.5
```

$ \Rightarrow gini(T,windy) = 0.429 $
